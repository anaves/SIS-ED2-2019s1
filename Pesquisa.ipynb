```python
def busca_sequencial(valor,lista):
    for e in lista:
        print(f'comparando {e} com buscado {valor}')
        if e == valor:
            return True # encontrei
    return False
```

```python
def busca_sequencial_ordenada(valor,lista):
    lista.sort() # ordena a lista
    print(lista)
    for e in lista:
        print(f'comparando {e} com buscado {valor}')
        if e == valor:
            return True # encontrei
        elif e > valor:
            break
    return False
```

```python
def busca_binaria(valor, lista):
    lista.sort()
    inicio = 0
    fim = len(lista)-1
    qtd = 0
    while inicio <= fim:
        meio = (inicio+fim)//2
       # print(f'Valor procurado {valor}')
        #print(lista[inicio:fim])
        qtd=qtd+1
        if lista[meio] == valor:
            print(f'QTD {qtd}')
            return True
        elif valor < lista[meio]:
            fim = meio - 1
        else:
            inicio = meio + 1
    else:
        print(f'QTD {qtd}')
        return False
```

```python
listaExemplo=[2,4,5,10,13,18,23,29,31,51,64]
```

```python
busca_binaria(10,listaExemplo)
```

```python
busca_binaria(12,listaExemplo)
```
